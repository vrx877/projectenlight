```python
import requests
from bs4 import BeautifulSoup
from IPython.display import clear_output

forbidden_links = [
    'mailto:studievejledningen@econ.ku.dk',
    'http://www.mozilla-europe.org/da/firefox/',
    'http://www.mozilla.com/en-US/firefox/ie.html',
    'http://www.adobe.com/products/acrobat/readstep2.html',
    'http://www.pkzip.com/',
    'http://www.download.com/'
]

def get_urls(url):
    html = requests.get(url).text
    soup = BeautifulSoup(html, "lxml")
    links = soup.find_all('a')
    if len(links)!=0:
        links = [str(x).split('\"')[1] for x in links] # Get HREF's
        links = [i for i in links if not i in forbidden_links] # Remove forbidden links
        links = [url+"?"+x.split('?')[1] if "?" in x else x for x in links] # Make absolute links
        links = [i.replace("æ","%E6") for i in links] # replace æ
        links = [i.replace("ø","%F8") for i in links] # replace ø
        links = [i.replace("å","%E5") for i in links] # replace å
        links = [i.replace("Æ","%C6") for i in links] # replace Æ
        links = [i.replace("Ø","%D8") for i in links] # replace Ø
        links = [i.replace("Å","%C5") for i in links] # replace Å
        links = [i.replace(" ","%20") for i in links] # replace space
        links = [i.replace("&amp;","&") for i in links] # replace &amp;
        return links
    else:
        return False

def savefile(url):
    response = requests.get(url,stream=True)
    if response.ok:
        f = open("assets/"+url.split("/")[-1], 'wb')
        f.write(response.content)
        f.close()
        return True
    else:
        return False
```

```python
start_url = "https://web3.econ.ku.dk/polit/studerende/eksamen/opgrv/cms-ku/dansk.asp"

fag = []
for url in get_urls(start_url):
    fag.extend(get_urls(url))

fag = ["fag_id="+str(x).split("&fag_id=")[1] for x in fag if "&fag_id=" in str(x)] # Split to only keep fag id and fag name
fag = list(set(fag)) # Remove duplicates

fag_ids = [x.split("&")[0][7:] for x in fag]
fag_names = [x.split("&")[1][8:] for x in fag]
fag_links = [start_url+"?"+x for x in fag]

fag_names = [i.replace("%E6","æ") for i in fag_names] # replace æ
fag_names = [i.replace("%F8","ø") for i in fag_names] # replace æ
fag_names = [i.replace("%E5","å") for i in fag_names] # replace æ
fag_names = [i.replace("%C6","Æ") for i in fag_names] # replace æ
fag_names = [i.replace("%D8","Ø") for i in fag_names] # replace æ
fag_names = [i.replace("%C5","Å") for i in fag_names] # replace æ
fag_names = [i.replace("%20"," ") for i in fag_names] # replace æ
```

```python
import pandas as pd
df = pd.DataFrame()
df['ID'] = fag_ids
df['ID'] = df['ID'].astype(int)
df["Name"] = fag_names
df["Link"] = fag_links
df = df.sort_values("ID")
df = df.reset_index(drop=True)

unique_fag = list(df["Link"])
```

```python
savename = "db2.csv"
all_titles = []
all_exams = []
all_answers = []
count = 0
for url in unique_fag:
    
    clear_output(wait=True)
    count=count+1
    
    html = requests.get(url).text
    table = html.split('<td class="politmain">')[1]
    soup = BeautifulSoup(table, "lxml")
    tables = soup.find_all("table")
    trs = tables[1].find_all("tr",height=False)
    
    titles = []
    exams = []
    answers = []
    
    for tr in trs:
        try: 
            title = tr.find_all("td")[0].text
        except: 
            title = "None"
        try:
            exam = tr.find_all("td")[1].find_all("a")[0]['href']
        except: 
            exam = "None"
                 
        try:
            answer = tr.find_all("td")[2].find_all("a")[0]['href']
        except:
            answer = "None"    
        
        titles.append(title)
        exams.append(exam)
        answers.append(answer)
    
    all_titles.append(titles)
    all_exams.append(exams)
    all_answers.append(answers)
    print("Scraped "+str(count)+"/"+str(len(unique_fag))+".")
    
df.to_csv(savename,sep=';',index=False)
print("Saved to "+savename+".")
```

```python
df2 = pd.DataFrame(columns=["ID","Name","Link","Exam"])
count_i = 0
count_j = 0
for i in all_titles:
    for j in i:
        fagID = df[df.index == count_i]["ID"].values[0]
        fagName = df[df.index == count_i]["Name"].values[0]
        fagLink = df[df.index == count_i]["Link"].values[0]
        df2 = df2.append({'Exam': j,"ID":fagID,"Name":fagName,"Link":fagLink}, ignore_index=True)
        #print(df.iloc[[count]]["ID"][0])
        count_j = count_j+1
    count_i=count_i+1
```

```python
df2["ID"] = df2["ID"].astype(int)
exams = []
for i in all_exams:
    for j in i:
        exams.append(j)
        
answers = []
for i in all_answers:
    for j in i:
        answers.append(j)
        
df2["Questions"] = exams
df2["Answers"] = answers
df2.to_csv("finaldb.csv",sep=';',index=False)
```
